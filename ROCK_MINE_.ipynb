```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.8, random_state=100)
```

### LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
Lr.fit(x_train,y_train)
LinearRegression
```

### RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

```python
rf_results = pd.DataFrame(['RANDOM FOREST REGRESSOR', None, None]).transpose()
rf_results.columns = ['METHOD', 'Training Data Accuracy', 'Test Data Accuracy']
```

### **LOGISTIC** **REGRESSION**

```python

from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train, y_train)
LogisticRegression()
```

```python
y_lr_train_pred= lr.predict(x_train)
y_lr_test_pred= lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
from sklearn.metrics import accuracy_score
training_data_accuracy= accuracy_score(y_lr_train_pred, y_train)
test_data_accuracy= accuracy_score(y_lr_test_pred, y_test)
```

```python
print("Accuracy score of training data:",training_data_accuracy)
```

```python
print("Accuracy score of test data:",test_data_accuracy)
```

```python
lr_results= pd.DataFrame(['LOGISTIC REGRESSION', training_data_accuracy, test_data_accuracy]).transpose()
lr_results.columns= ['METHOD', 'Training Data Accuracy', 'Test Data Accuracy']
```
